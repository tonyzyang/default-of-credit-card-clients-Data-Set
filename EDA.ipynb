```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel("default of credit card clients.xls", header = 1)
```

```python
zero = df.EDUCATION == 0
df.EDUCATION = df.EDUCATION.mask(zero, np.nan)
zero1 = df.MARRIAGE == 0
df.MARRIAGE = df.MARRIAGE.mask(zero1, np.nan)
df = df.dropna()
```
